```python
from csdid.att_gt import ATTgt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
data_final = pd.read_csv("../output/quadkeys_final_df.csv")
data_final.head()
data_final.columns
import matplotlib.pyplot as plt

# Calculate the 99th percentile
percentile_99 = data_final['avg_d_mbps'].quantile(0.99)

# Print the 99th percentile
print(f"99th Percentile of avg_d_mbps: {percentile_99}")

# Plot the histogram
data_final['avg_d_mbps'].hist(bins=50, edgecolor='black')
plt.axvline(percentile_99, color='red', linestyle='dashed', linewidth=1, label=f"99th Percentile ({percentile_99:.2f})")
plt.title('Histogram of avg_d_mbps')
plt.xlabel('avg_d_mbps')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Calculate the 99th percentile
percentile_99 = data_final['avg_d_mbps'].quantile(0.999)

# Filter the DataFrame for values greater than the 99th percentile
values_greater_than_99th = data_final[data_final['avg_d_mbps'] > percentile_99]

# Display the filtered values
values_greater_than_99th.avg_d_mbps.hist()
```
